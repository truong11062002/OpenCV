```
#bai toan 2
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/123689546_1017193625448751_8960630193307838067_n.jpg', 0)

hist1 , bins = np.histogram(img.ravel(),256,[0,256])
hist2 = np.bincount(img.ravel(),minlength=256)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.plot(hist1)
plt.title('np.histogram')

plt.subplot(1,3,3) ,plt.plot(hist2)
plt.title('np.bincount')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/espresso_color.jpg', 1)

img1 = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,6))
plt.subplot(1,3,1), plt.imshow(img1 ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2)
color = ('b','g','r')
for i,col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
plt.show()  
```

```
#bai toan 4
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/hist_1.PNG', 0)
img1 = cv2.equalizeHist(img)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 ,cmap='gray')
plt.title('Histogram Equalization')
plt.show()
```
