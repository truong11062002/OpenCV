```
#bai toan 2
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/lindsay.tif',0 )
height , width = img.shape

trX = 60
trY = 30

M = np.float32([[1,0,trX] , [0,1,trY]])
img1 = cv2.warpAffine(img , M,(width,height))
img2 = cv2.warpAffine(img1 , M,(width,height))

plt.figure(figsize=(12,12))

plt.subplot(1,3,1) , plt.imshow(img , cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) , plt.imshow(img1 ,cmap ='gray')
plt.title('Translation x = {} y = {}'.format(trX , trY))

plt.subplot(1,3,3) , plt.imshow(img2 ,cmap ='gray')
plt.title('Translation x = {} y= {} '.format(trX , trY))
```

```
#bai toan 3
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/lindsay.tif',0 )
height , width = img.shape

M = cv2.getRotationMatrix2D((height/2,width/2),20,1)
img1 = cv2.warpAffine(img , M, (width,height))
M = cv2.getRotationMatrix2D((height/2,width/2),30,1)
img2 = cv2.warpAffine(img1 , M, (width,height))

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Rotation angle: pi/9')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Rotation angle: pi/9+pi/6')
```
