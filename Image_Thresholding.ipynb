```
#Bai toan 3
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/gradient.tif',0)

ret1, img1 = cv2.threshold(img,127,255, cv2.THRESH_TOZERO)
ret2, img2 = cv2.threshold(img,127,255, cv2.THRESH_TOZERO_INV)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1) , plt.imshow(img , cmap ='gray')
plt.title('Original')

plt.subplot(1,3,2) , plt.imshow(img1 , cmap=('gray'))
plt.title('THRESH_TOZERO')

plt.subplot(1,3,3) , plt.imshow(img2 , cmap=('gray'))
plt.title('THRESH_TOZERO_INV')
```

```
#Bai toan 4
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/gradient_with_text.tif',0)

ret1, img1 = cv2.threshold(img,127,255, cv2.THRESH_BINARY)

img2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9 ,2)
img3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9 ,2)

plt.figure(figsize=(12,12))
plt.subplot(2,2,1) , plt.imshow(img , cmap ='gray')
plt.title('Original')

plt.subplot(2,2,2) , plt.imshow(img1 , cmap=('gray'))
plt.title('THRESH_BINARY')

plt.subplot(2,2,3) , plt.imshow(img2 , cmap=('gray'))
plt.title('THRESH_MEAN_C')

plt.subplot(2,2,4) , plt.imshow(img3 , cmap=('gray'))
plt.title('THRESH_GAUSSIAN_C')
```
