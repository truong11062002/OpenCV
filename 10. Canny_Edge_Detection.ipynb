```
#bai toan 1
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/lenna.tif', 0)
img1 = cv2.GaussianBlur(img ,(5,5),0)
img2 = cv2.GaussianBlur(img ,(17,17),0)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Gaussian Filter 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Gaussian Filter 17x17')
```

```
#bai toan 3
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/lenna.tif', 0)
img1 = cv2.GaussianBlur(img , (5,5), 0)
img2 = cv2.Canny(img , 100, 200)
img3 = cv2.Canny(img1 , 100, 200)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img2 , cmap='gray')
plt.title('Canny edge')

plt.subplot(1,3,3) ,plt.imshow(img3 , cmap='gray')
plt.title('Gaussian Filter 5x5 + Canny edge')
```
