```
#bai toan 1
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/eight_pepper.tif', 0)
kernel = np.ones((3,3) , np.float32)/9
img1 = cv2.filter2D(img , -1,kernel)
kernel = np.ones((5,5),np.float32)/25
img2 = cv2.filter2D(img , -1,kernel)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Mean Filter 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Mean Filter 5x5')
```

```
#bai toan 2
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/eight_pepper.tif', 0)
img1 = cv2.blur(img ,(3,3))
img2 = cv2.blur(img ,(5,5))

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Blur 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Blur 5x5')
```

```
#bai toan 3
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/eight_pepper.tif', 0)
img1 = cv2.GaussianBlur(img ,(3,3) ,0)
img2 = cv2.GaussianBlur(img ,(5,5) ,0)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Gaussian 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Gaussian 5x5')
```
