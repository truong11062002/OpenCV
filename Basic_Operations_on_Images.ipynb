```
#Bai toan 6
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/statue.png',1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(img)

R,G,B = cv2.split(img)
img = cv2.merge((R,G,B))
plt.subplot(1,2,1) , plt.imshow(img)
plt.title('Merge chanels R,G,B')
```
