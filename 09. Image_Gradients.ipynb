```
#bai toan 2
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/sudoku.PNG', 0)
img1 = cv2.Laplacian(img ,cv2.CV_64F , ksize = 3)
img2 = cv2.Laplacian(img ,cv2.CV_64F , ksize = 9)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Laplation 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Laplation 9x9')
```

```
#bai toan 1
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/sudoku.PNG', 0)
img1 = cv2.Sobel(img ,cv2.CV_64F,1,0 , ksize = 3)
img2 = cv2.Sobel(img ,cv2.CV_64F,0,1 , ksize = 9)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Sobel 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Sobel 9x9')
```
