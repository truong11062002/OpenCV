```
import matplotlib.pyplot as plt
import numpy as np
import cv2
from numpy import random

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/salzburg.png', 1)
#rectangle(left,top,right,button)

l = random.randint(img.shape[1]/2)
t = random.randint(img.shape[0]/2)
r = random.randint(img.shape[1]/2,img.shape[1])
b = random.randint(img.shape[0]/2,img.shape[0])

img = cv2.rectangle(img,(l,t),(r,b),(0,255,0),5)
plt.figure(figsize=(6,6))
plt.imshow(img)
```

```
import matplotlib.pyplot as plt
import numpy as np
import cv2
from numpy import random

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/salzburg.png', 1)
radius = random.randint(50,100)

x = random.randint(img.shape[1]/4,img.shape[1]/2)
y = random.randint(img.shape[0]/4,img.shape[0]/2)

img = cv2.circle(img , (x,y) , 10 ,(255,0,0) , -1)
img = cv2.circle(img , (x,y) , radius,(0,255,0) , 5)
plt.figure(figsize=(6,6))
plt.title('Original')
plt.imshow(img)
print(img.shape)
```

```
import matplotlib.pyplot as plt
import numpy as np
import cv2
from numpy import random

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/salzburg.png', 1)

x = random.randint(img.shape[1]/4,img.shape[1]/2)
y = random.randint(img.shape[0]/4,img.shape[0]/2)

axesLength = (150,100)
angle = 30
startAngle = 0
endAngle = 360
color =( 0,255,0)
thickness = -1

img = cv2.ellipse(img,(x,y), axesLength,angle,startAngle,endAngle,color,thickness)
plt.figure(figsize=(6,6))
plt.imshow(img)
print(img.shape)
```

```
import matplotlib.pyplot as plt
import numpy as np
import cv2
from numpy import random

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/salzburg.png', 1)

pts = np.array([[100,100],[600,100],[400,500],[200,500]])
pts = pts.reshape((pts.shape[0],1,pts.shape[1]))

img = cv2.polylines(img,[pts],True,(255,0,0),10)
plt.figure(figsize=(6,6))
plt.title('Original')
plt.imshow(img)
print(img.shape)
```

```
import matplotlib.pyplot as plt
import numpy as np
import cv2
from numpy import random

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/salzburg.png', 1)

x = random.randint(img.shape[1]/4,img.shape[1]/2)
y = random.randint(img.shape[0]/4,img.shape[0]/2)

font = cv2.FONT_HERSHEY_COMPLEX
text ='OpenCV'
org =(x,y)
fontScale = 4
color =(0,255,0)
thickness = 5
plt.figure(figsize=(6,6))
img = cv2.putText(img,text,org,font,fontScale,color,thickness,cv2.LINE_AA)
plt.imshow(img, interpolation= 'bicubic')
plt.title('Original')
```
