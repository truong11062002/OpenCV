```
#bai toan 1
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('/content/drive/MyDrive/00. Dataset Image Processing/morph.bmp' ,0)
kernel = np.ones((3,3) ,np.uint8)
img1 = cv2.erode(img ,kernel , iterations =1)
kernel = np.ones((9,9) ,np.uint8)
img2 = cv2.erode(img ,kernel , iterations =1)

plt.figure(figsize=(12,12))
plt.subplot(1,3,1), plt.imshow(img ,cmap='gray')
plt.title('Original')

plt.subplot(1,3,2) ,plt.imshow(img1 , cmap='gray')
plt.title('Erosion 3x3')

plt.subplot(1,3,3) ,plt.imshow(img2 , cmap='gray')
plt.title('Erosion 5x5')
```
